```python
import numpy as np
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
```

```python
data=load_boston()
```

```python
#functions modified for generic gradient descent
```

```python
def step_gradient(X, Y, m, learning_rate):
    m_slope=np.zeros(len(X[0]))
    for i in range(len(X)):
        x=X[i]
        y=Y[i]
        for j in range(len(x)):
            m_slope[j]+=(-2/len(x))*(y-sum(m*x))*x[j]
    new_m=m-(learning_rate*m_slope)
    return new_m
```

```python
def cost(m, x, y):
    cost=0
    for i in range(len(x)):
        cost+=((y[i]-sum(m*x[i]))**2)
    print(cost/(len(x)))
```

```python
def gd(x, y, learning_rate, iterations):
    m=np.zeros(len(x[0]))
    for i in range(iterations):
        m=step_gradient(x, y, m, learning_rate)
        cost(m, x, y)
    return m
```

```python
def run():
    iterations=300
    learning_rate=0.00000001
    x=np.append(data.data, np.ones(len(data.data)).reshape(-1, 1), axis=1)
    y=data.target
    m=gd(x, y, learning_rate, iterations)
run()
```
